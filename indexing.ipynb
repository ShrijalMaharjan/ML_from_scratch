##Group by and Aggregarting

```python
import pandas as pd
df = pd.read_csv("/content/Flavors.csv")
df
```

```python
group_by_frame = df.groupby('Base Flavor')
```

```python
group_by_frame.mean(numeric_only=True)
```

```python
group_by_frame.count()
```

```python
group_by_frame.sum()
```

```python
df.groupby(['Base Flavor', 'Liked']).agg({'Flavor Rating': ['mean', 'max', 'count', 'sum']})
```

##Data merge, concatinate and Join

```python
import pandas as pd
df1 = pd.read_csv("/content/LOTR.csv")
df2 = pd.read_csv("/content/LOTR 2.csv")
```

```python
df1
```

```python
df2
```

```python
df1.merge(df2, how='inner', on=['FellowshipID', 'FirstName'])
```

```python
df1.merge(df2, how='outer')
```

```python
df1.merge(df2, how='left')
```

```python
df1.merge(df2, how='right')
```
